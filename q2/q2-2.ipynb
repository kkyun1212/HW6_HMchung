```python
def main():
    import csv
    import matplotlib.pyplot as plt

    data100=[]
    data1000=[]
    data10000=[]
    data100000=[]
    
    f = open('write.csv','r',encoding='ANSI')
    data = csv.reader(f)

    for row in data:
        if float(row[0])<101:
            data100.append(float(row[1]))
        elif float(row[0])<1101:
            data1000.append(float(row[1]))
        elif float(row[0])<11101:
            data10000.append(float(row[1]))
        else:
            data100000.append(float(row[1]))
    
    f.close()

    plt.title('random number 100 times')
    plt.hist(data100)
    plt.xlabel('number')
    plt.ylabel('times')
    plt.show()
    
    plt.title('random number 1000 times')
    plt.hist(data1000)
    plt.xlabel('number')
    plt.ylabel('times')
    plt.show()
    
    plt.title('random number 10000 times')
    plt.hist(data10000)
    plt.xlabel('number')
    plt.ylabel('times')
    plt.show()
    
    plt.title('random number 100000 times')
    plt.hist(data100000)
    plt.xlabel('number')
    plt.ylabel('times')
    plt.show()
    
if __name__ == '__main__':
    main()
```
