```python
def main():
    import csv
    import matplotlib.pyplot as plt

    male=[]
    female=[]
    years=[2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]
    n=0

    f = open('jejuss.csv','r',encoding='ANSI')
    data = csv.reader(f)
    
    header = next(data)

    for row in data:
        while n<10:
            male.append(float(row[4*n+2]))
            female.append(float(row[4*n+3]))
            n=n+1
    
    
    f.close()

    plt.title('jeju')
    
    plt.xlabel('year')
    plt.ylabel('number of people')
    plt.plot(years,male,label='male')
    plt.plot(years,female,label='female')
    plt.legend()
    plt.show()
    
if __name__ == '__main__':
    main()
```
