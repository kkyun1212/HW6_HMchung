```python
def main():
    import csv
    import operator
    import matplotlib.pyplot as plt
    
    plt.rcParams['font.family'] ='Malgun Gothic'
    plt.rcParams['axes.unicode_minus'] =False

    with open('subway.csv', 'r', encoding='UTF-8') as f:
        reader = csv.reader(f)
        data = [row for row in reader]

    ride = sorted(data, key=lambda x: int(x[4])+int(x[6]), reverse=True)[:30]
    stations = [row[3] for row in ride]
    counts = [int(row[4])+int(row[6]) for row in ride]

    plt.bar(stations, counts)
    plt.title('     30')
    plt.xlabel('')
    plt.ylabel('')
    plt.xticks(rotation=90)
    plt.show()

    exit = sorted(data, key=lambda x: int(x[5])+int(x[7]), reverse=True)[:30]
    stations = [row[3] for row in exit]
    counts = [int(row[5])+int(row[7]) for row in exit]

    plt.bar(stations, counts)
    plt.title('     30')
    plt.xlabel('')
    plt.ylabel('')
    plt.xticks(rotation=90)
    plt.show()

    rideExit = sorted(data, key=lambda x: int(x[4])+int(x[5])+int(x[6])+int(x[7]), reverse=True)[:30]
    stations = [row[3] for row in rideExit]
    counts = [int(row[4])+int(row[5])+int(row[6])+int(row[7]) for row in rideExit]

    plt.bar(stations, counts)
    plt.title('     30')
    plt.xlabel('')
    plt.ylabel('')
    plt.xticks(rotation=90)
    plt.show()


if __name__ == '__main__':
    main()
```
